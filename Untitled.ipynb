```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1):
        self.weights = np.random.randn(input_size)
        self.bias = np.random.randn()
        self.learning_rate = learning_rate
        
    def forward(self, inputs):
        # Linear combination: wx + b
        self.inputs = inputs
        self.linear_output = np.dot(inputs, self.weights) + self.bias
        
        # Activation function (step function for classification)
        self.output = 1 if self.linear_output >= 0 else 0
        return self.output
    
    def update(self, error):
        # Update weights and bias based on error
        self.weights += self.learning_rate * error * self.inputs
        self.bias += self.learning_rate * error

# Create a perceptron that learns AND logic
def train_perceptron(logic_gate):
    # Training data for AND gate
    X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    if logic_gate == "AND":
        y = np.array([0, 0, 0, 1])
    elif logic_gate == "OR":
        y = np.array([0, 1, 1, 1])
    
    # Create and train perceptron
    perceptron = Perceptron(input_size=2, learning_rate=0.1)
    
    errors = []
    for epoch in range(100):
        total_error = 0
        for i in range(len(X)):
            prediction = perceptron.forward(X[i])
            error = y[i] - prediction
            perceptron.update(error)
            total_error += abs(error)
        errors.append(total_error)
        
        # Stop early if converged
        if total_error == 0:
            break
    
    # Test the trained perceptron
    print(f"\n{logic_gate} Gate Results:")
    for i in range(len(X)):
        prediction = perceptron.forward(X[i])
        print(f"Input: {X[i]}, Prediction: {prediction}, Expected: {y[i]}")
    
    return perceptron, errors

# Train AND and OR perceptrons
and_perceptron, and_errors = train_perceptron("AND")
or_perceptron, or_errors = train_perceptron("OR")

# Plot training progress
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(and_errors)
plt.title('AND Gate Training Error')
plt.xlabel('Epoch')
plt.ylabel('Total Error')

plt.subplot(1, 2, 2)
plt.plot(or_errors)
plt.title('OR Gate Training Error')
plt.xlabel('Epoch')
plt.ylabel('Total Error')
plt.tight_layout()
plt.show()
```
